```python
# Task 2 - Simple Text Classifier
def gender_features(word): #gives last letter of the word
    return {'last_letter':word[-1]}
```

```python
#giving names
gender_features('Pratik')
```

```python
from nltk.corpus import names
labeled_names = ([(name, 'male') for name in names.words('male.txt')]+
[(name, 'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names)
```

```python
featuresets = [(gender_features(n),gender) for (n,gender) in labeled_names] #this is used to train the model
#this will have last letter of names followed by gender
```

```python
featuresets
```

```python
train_set,test_set=featuresets[500:],featuresets[:500]
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
import nltk
classifier2 =nltk.DecisionTreeClassifier.train(train_set)
```

```python
classifier.classify(gender_features('Pratik'))
```

```python
classifier2.classify(gender_features('Pratik'))
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

```python
print(nltk.classify.accuracy(classifier2, test_set))
```

## Task 3 - Vectorisers & cosine Similarity

```python
#read more about countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
#from sklearn.feature_extraction.text import TfidVectorizer
```

```python
vect = CountVectorizer(binary = True)
corpus = ["Tessaract is good optical character recognition engine  ", "optical character recognition is significant "]
vect.fit(corpus) #it is the most imp step we need to fit the corpus to vectorizer
```

```python
vocab=vect.vocabulary_ #built-in function where index for unique words in corpus
```

```python
for key in sorted(vocab.keys()): #indexing of the words from corpus
     print("{}:{}".format(key, vocab[key]))
```

```python
print(vect.transform(["This is a good optical illusion"]).toarray()) #tranform helps to do with any sentence other than in corpus
#we took one sentence and vectorized using vet vectorizer and got that 3 words are same
```

```python
print(vect.transform(corpus).toarray())
```

```python
#cosine similartiy works, we can have different distance and similarity formula
from sklearn.metrics.pairwise import cosine_similarity
```

```python
print(vect.transform(["Google Cloud Vision is a character recognition engine"]).toarray())
print(vect.transform(["OCR is an optical character recognition engine"]).toarray())
```

```python
similarity = cosine_similarity(vect.transform(["Google Cloud Vision is a character recognition engine"]).toarray(), vect.transform(["OCR is an optical character recognition engine"]).toarray()) #two different text given to vet and then cosine similarity 
```

```python
print(similarity) #similarity between this two sentences
```
